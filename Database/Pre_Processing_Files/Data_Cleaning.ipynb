```python
# Initial imports
import pandas as pd
from pathlib import Path
```

```python
# Load the crypto_data.csv dataset.
file_path = "Database/New_CSV_Files/NBA_GAMESDATA.csv"
df = pd.read_csv(file_path)
df.head(10)
```

```python
#Drop unnecessary columns
df = df.drop(['TEAM_ABBREVIATION', 'TEAM_NAME', 'MIN'], axis=1)
# Set Winning & Losing Columns
df['WIN'] = ['1' if x=='W' else '0' for x in df['WL']]
```

```python
# Home Team has @ sign after name, so split match up column into three columns
# Home team have rows with @ symbol and away teams do not
df[['Team_1', 'symbol', 'Team_2']] = df.MATCHUP.str.split(expand=True)
```

```python
df_home = df.loc[df['symbol'] == '@']
df_away = df.loc[df['symbol'] != '@']
```

```python
#Change header names to HOME_[Column Name]
df_home.rename(columns={'TEAM_ID': 'HOME_TEAM_ID', 
                       'TEAM_NAME': 'HOME_TEAM_NAME', 
                        'PTS': 'HOME_PTS',
                        'FGM': 'HOME_FGM', 
                        'FGA': 'HOME_FGA',
                        'FG_PCT': 'HOME_FG_PCT', 
                        'FG3M': 'HOME_FG3M',
                        'FG3A': 'HOME_FG3A', 
                        'FG3_PCT': 'HOME_FG3_PCT',
                        'FTM': 'HOME_FTM', 
                        'FTA': 'HOME_FTA',
                        'FT_PCT': 'HOME_FT_PCT', 
                        'OREB': 'HOME_OREB',
                        'DREB': 'HOME_DREB',
                        'REB': 'HOME_REB',
                        'AST': 'HOME_AST',
                        'STL': 'HOME_STL',
                        'BLK': 'HOME_BLK',
                        'TOV': 'HOME_TOV',
                        'PF': 'HOME_PF',
                        'PLUS_MINUS': 'HOME_PLUS_MINUS',
                        'WIN': 'HOME_WIN'}, 
               inplace=True)

#Change header names to AWAY_[Column Name]
df_away.rename(columns={'TEAM_ID': 'AWAY_TEAM_ID',
                       'TEAM_NAME': 'AWAY_TEAM_NAME', 
                        'PTS': 'AWAY_PTS',
                        'FGM': 'AWAY_FGM', 
                        'FGA': 'AWAY_FGA',
                        'FG_PCT': 'AWAY_FG_PCT', 
                        'FG3M': 'AWAY_FG3M',
                        'FG3A': 'AWAY_FG3A', 
                        'FG3_PCT': 'AWAY_FG3_PCT',
                        'FTM': 'AWAY_FTM', 
                        'FTA': 'AWAY_FTA',
                        'FT_PCT': 'AWAY_FT_PCT', 
                        'OREB': 'AWAY_OREB',
                        'DREB': 'AWAY_DREB',
                        'REB': 'AWAY_REB',
                        'AST': 'AWAY_AST',
                        'STL': 'AWAY_STL',
                        'BLK': 'AWAY_BLK',
                        'TOV': 'AWAY_TOV',
                        'PF': 'AWAY_PF',
                        'PLUS_MINUS': 'AWAY_PLUS_MINUS',
                        'WIN': 'AWAY_WIN'}, 
               inplace=True)
```

```python
df_home.head()
df_away.head()
```

```python
#Drop unnecessary home team columns
df_home = df_home.drop(['MATCHUP', 'WL', 'Team_1', 'symbol', 'Team_2'], axis=1)
#Drop unnecessary away team columns
df_away = df_away.drop(['MATCHUP', 'WL', 'Team_1', 'symbol', 'Team_2'], axis=1)
```

```python
df2 = df_home.merge(df_away, how = 'outer', on = 'GAME_ID')
```

```python
df2.info()
```

```python
#Drop unnecessary merged columns
df2 = df2.drop(['GAME_DATE_y', 'SEASON_ID_y', 'AWAY_WIN'], axis=1)
#Drop NA Values
df2 = df2.dropna()
df2
```

```python
#Rename SEASON_ID and GAME_DATE columns
df2.rename(columns={'GAME_DATE_x': 'GAME_DATE', 
                       'SEASON_ID_x': 'SEASON_ID'}, 
               inplace=True)
```

```python
df2['GAME_DATE'] = df['GAME_DATE'].astype('datetime64[D]')
df2.info()
```

```python
df2.drop_duplicates(subset=['GAME_ID'], keep='first')
df2["HOME_WIN"] = df2["HOME_WIN"].astype(str).astype(int)
df2.info()
```
